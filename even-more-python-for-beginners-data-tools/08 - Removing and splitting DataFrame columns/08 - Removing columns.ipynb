## DataFrame에서 열 지우기

기계 학습을 위해 데이터를 준비 할 때 모델을 학습하기 전에 DataFrame에서 특정 열을 삭제해야 할 수 있습니다.

예시:
비행 지연 시간 (ARR_DELAY)을 예측하기 위해 모델을 학습한다고 가정 해보십시오.

모델이 예정된 도착 시간 (CRS_ARR_TIME)과 실제 도착 시간 (ARR_TIME)을 알고있는 경우 모델은 ARR_DELAY = ARR_TIME-CRS_ARR_TIME을 신속하게 파악합니다.

향후 항공편의 도착 시간을 예측할 때 도착 시간 (ARR_TIME) 값이 없습니다. 따라서 ARR_DELAY를 예측하도록 모델을 훈련 할 때 추가로 사용되지 않도록 DataFrame에서 이 열을 제거해야합니다.

```python
# Remove the column ARR_TIME from the DataFrane delays_df

#delays_df = delays_df.drop(['ARR_TIME'],axis=1)
new_df = delays_df.drop(columns=['ARR_TIME'])
new_df.head()
```

** inplace ** 매개 변수를 사용하여 원래 DataFrame에서 열을 삭제하도록 지정합니다

```python
# Remove the column ARR_TIME from the DataFrame delays_df

#delays_df = delays_df.drop(['ARR_TIME'],axis=1)
delays_df.drop(columns=['ARR_TIME'], inplace=True)
delays_df.head()
```

우리는 보통 숫자 값(예 : 거리, 분, 무게)나 숫자가 아닌 정량적인 값(예 : 항공편이 출발한 공항, 항공편을 운항한 항공사)을 기반으로 예측하기 위해 다양한 기술을 사용합니다.

정량적 데이터는 모델을 학습하기 전에 별도의 DataFrame으로 이동할 수 있습니다.

그리고 예측을 수행하는데 도움이 될 수 있다고 생각되는 값(특징)과는 별도의 DataFrame에 ARR_DELAY 레이블이라는 예측하려는 값을 넣어야합니다.

기존 데이터 프레임의 열에서 새 데이터 프레임을 생성할 수 있어야합니다.

```python
# desc_df의 이름을 가지는 DataFrame을 생성하세요
# 모든 행을 포함하세요.
# ORIGIN, DEST, OP_CARRIER_FL_NUM, OP_UNIQUE_CARRIER, TAL_NUM 열을 포함하세요.

desc_df = delays_df.loc[:,['ORIGIN', 'DEST', 'OP_CARRIER_FL_NUM', 'OP_UNIQUE_CARRIER', 'TAIL_NUM']]
desc_df.head()
```
