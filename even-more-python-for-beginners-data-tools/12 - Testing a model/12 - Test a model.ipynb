# 학습된 모델 테스트하기
학습된 모델을 가지고 있다면 추가적인 데이터를 이용해 테스트 할 수 있습니다.

학습된 모델을 생성하기 위해 이전 notebook의 코드를 재실행하는 것을 시작할 것입니다.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# csv 파일로부터 데이터 불러오기
delays_df = pd.read_csv('Data/Lots_of_flight_data.csv') 

# null 값이 존재하면 선형 회귀 모델을 테스트하는 것을 망칠 수 있기 때문에 null 값이 있는 행을 제거해줍니다.
delays_df.dropna(inplace=True)

# X DataFrame으로 특성 옮기기
X = delays_df.loc[:,['DISTANCE', 'CRS_ELAPSED_TIME']]

# y DataFrame으로 레이블 옮기기
y = delays_df.loc[:,['ARR_DELAY']] 

# 테스트용,학습용 DataFrame으로 데이터 나누기
X_train, X_test, y_train, y_test = train_test_split(
                                                    X, 
                                                    y, 
                                                    test_size=0.3, 
                                                    random_state=42
                                                   )
regressor = LinearRegression()     # scikit learn의 LinearRegression 객체를 이용하여 생성
regressor.fit(X_train, y_train)    # 학습용 데이터 사용해 모델 학습하기 위해 fit 메소드 이용
```

## 모델 테스트
테스트 데이터로 학습된 모델의 예측값을 얻기 위해 **Scikitlearn LinearRegression predict** 를 이용합니다.

우리는 X_test에 테스트용 데이터를 담고있습니다.

예측된 값을 y_pred에 저장합니다.

```python
y_pred = regressor.predict(X_test)
```
